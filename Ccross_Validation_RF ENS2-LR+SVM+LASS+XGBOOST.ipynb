```python
import numpy as np
import pandas as pd
```

```python

# Logistic Regression
lr = LogisticRegression(solver='liblinear', random_state=42)
```

```python
from sklearn.pipeline import make_pipeline
```

```python
# SVM with Lasso (Lasso is used as feature selection within a pipeline)
svm_lasso = make_pipeline(StandardScaler(), SVC(kernel='linear', probability=True, random_state=42))
```

```python
# XGBoost Classifier
xgb = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
# SVM with L1 regularization (LASSO)
svm = SVC(probability=True)  # Set probability=True for soft voting
svm_lasso = Pipeline([
    ('scaler', StandardScaler()),  # SVM usually benefits from feature scaling
    ('svm', SVC(kernel='linear', probability=True))
])
```

```python

# XGBoost
xgb = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
```

```python
# Hyperparameter grids
param_grid_lr = {
    'C': [0.1, 1, 10],
    'solver': ['liblinear', 'lbfgs'],
    'penalty': ['l2'],
    'max_iter': [100, 200]
}

param_grid_svm = {
    'svm__C': [0.1, 1, 10]
}

param_grid_xgb = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0]
}
```

```python

# Grid search for each model
grid_lr = GridSearchCV(lr, param_grid_lr, cv=5, scoring='accuracy')
grid_svm = GridSearchCV(svm_lasso, param_grid_svm, cv=5, scoring='accuracy')
grid_xgb = GridSearchCV(xgb, param_grid_xgb, cv=5, scoring='accuracy')
```

```python
# Fit the models
grid_lr.fit(X_train, y_train)
grid_svm.fit(X_train, y_train)
grid_xgb.fit(X_train, y_train)
```

```python
# Best models
best_lr = grid_lr.best_estimator_
best_svm = grid_svm.best_estimator_
best_xgb = grid_xgb.best_estimator_
```

```python
# Ensemble with soft voting
voting_clf = VotingClassifier(estimators=[
    ('lr', best_lr),
    ('svm', best_svm),
    ('xgb', best_xgb)
], voting='soft')
```

```python
from sklearn.model_selection import cross_val_score
```

```python
# Cross-validation for ensemble
cv_scores = cross_val_score(voting_clf, X_train, y_train, cv=5, scoring='accuracy')
print(f"Cross-validation accuracy scores: {cv_scores}")
print(f"Mean cross-validation accuracy: {cv_scores.mean()}")
```

```python
# Train the ensemble on the full training data
voting_clf.fit(X_train, y_train)
```

```python
# Make predictions
y_pred = voting_clf.predict(X_test)
y_train_pred=voting_clf.predict(X_train)
```

```python
train_accuracy_voting_clf = round(accuracy_score(y_train_pred, y_train), 3)
accuracy_voting_clf = round(accuracy_score(y_pred, y_test), 3)
precision_score_voting_clf = round(precision_score(y_pred, y_test), 3)
recall_score_voting_clf = round(recall_score(y_pred, y_test), 3)
f1_score_voting_clf = round(f1_score(y_pred, y_test), 3)
auc_voting_clf = round(roc_auc_score(y_pred,y_test), 3)

print("The accuracy on train data is ", train_accuracy_voting_clf)
print("The accuracy on test data is ", accuracy_voting_clf)
print("The precision on test data is ", precision_score_voting_clf)
print("The recall on test data is ", recall_score_voting_clf)
print("The f1 on test data is ", f1_score_voting_clf)
print("The auc on test data is ", auc_voting_clf)
```
